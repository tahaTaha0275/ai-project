### Since our dataset has a the problem of imbalanced data  we are going to use stratified k-fold in order to perform cross validation

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

data = pd.read_csv('./heart_disease_uci_formatted_and_cleaned.csv')

display(data)

columns_to_encode = ['sex', 'fbs', 'exang']

encoder = LabelEncoder()

for col in columns_to_encode:
    data[col] = encoder.fit_transform(data[col])

data = pd.get_dummies(data, columns=['cp'], dtype=int)
data = pd.get_dummies(data, columns=['restecg'], dtype=int)
data = pd.get_dummies(data, columns=['slope'], dtype=int)
data = pd.get_dummies(data, columns=['thal'], dtype=int)


data['num'] = data.pop('num')
display(data)
data['num'].value_counts()
```

```python
from sklearn.model_selection import StratifiedKFold
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense, LeakyReLU
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import numpy as np
from imblearn.over_sampling import SMOTE
import tensorflow as tf
from sklearn.utils.class_weight import compute_class_weight
import matplotlib.pyplot as plt
import seaborn as sns

# Set up Stratified K-Folds cross-validator
k_folds = 20
skf = StratifiedKFold(n_splits=k_folds, shuffle=True, random_state=42)
smote = SMOTE(random_state=42)

# To store performance metrics for each fold
fold_accuracies = []
fold_precisions = []
fold_recalls = []
fold_f1_scores = []

# Initialize cumulative confusion matrix
num_classes = 5  # Number of classes (adjust if needed)
cumulative_conf_matrix = np.zeros((num_classes, num_classes), dtype=int)

# Model creation function
def create_model():
    model = Sequential([
        Input(shape=(21,)),             # Input layer with 21 features
        Dense(64, activation='relu'),   # First hidden layer
        Dense(32, activation='relu'),   # Second hidden layer
        Dense(32),                       
        LeakyReLU(negative_slope=0.01),
        Dense(5, activation='softmax')  # Output layer for 5 classes
    ])
    model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

# Create a new model for each fold to reset weights
model1 = create_model()

# Load data (ensure `data` is defined earlier in your code)
X = data.iloc[:, :-1]
y = data.iloc[:, -1]

# Loop through each fold
for fold, (train_index, val_index) in enumerate(skf.split(X, y)):
    X_train_fold, X_val_fold = X.iloc[train_index], X.iloc[val_index]
    y_train_fold, y_val_fold = y.iloc[train_index], y.iloc[val_index]
    
    # Apply SMOTE to balance the dataset
    X_train_fold, y_train_fold = smote.fit_resample(X_train_fold, y_train_fold)

    X_train_fold = np.array(X_train_fold)
    y_train_fold = np.array(y_train_fold)
    X_val_fold = np.array(X_val_fold)
    y_val_fold = np.array(y_val_fold)

    # Calculate class weights
    class_weights = compute_class_weight('balanced', classes=np.unique(y_train_fold), y=y_train_fold)
    class_weights_dict = {i: weight for i, weight in enumerate(class_weights)}

    # Train the model with class weights
    model1.fit(X_train_fold, y_train_fold, epochs=200, batch_size=16, 
               verbose=0, class_weight=class_weights_dict)

    # Get predictions
    y_pred = np.argmax(model1.predict(X_val_fold), axis=1)

    # Calculate metrics
    accuracy = accuracy_score(y_val_fold, y_pred)
    precision = precision_score(y_val_fold, y_pred, average='macro', zero_division=1)
    recall = recall_score(y_val_fold, y_pred, average='macro')
    f1 = f1_score(y_val_fold, y_pred, average='macro')
    
    # Append metrics
    fold_accuracies.append(accuracy)
    fold_precisions.append(precision)
    fold_recalls.append(recall)
    fold_f1_scores.append(f1)

    # Compute confusion matrix for the fold and add to cumulative matrix
    fold_conf_matrix = confusion_matrix(y_val_fold, y_pred, labels=np.arange(num_classes))
    cumulative_conf_matrix += fold_conf_matrix

    # Print metrics for each fold
    print(f"Fold {fold + 1} - Accuracy: {accuracy:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}, F1-score: {f1:.4f}")

# Average metrics across all folds
avg_accuracy = np.mean(fold_accuracies)
avg_precision = np.mean(fold_precisions)
avg_recall = np.mean(fold_recalls)
avg_f1_score = np.mean(fold_f1_scores)

print(f"\nAverage metrics across {k_folds} folds:")
print(f"Accuracy: {avg_accuracy:.4f}")
print(f"Precision: {avg_precision:.4f}")
print(f"Recall: {avg_recall:.4f}")
print(f"F1-score: {avg_f1_score:.4f}")

# Final predictions on the full dataset
y_pred = np.argmax(model1.predict(np.array(X)), axis=1)

# Calculate final metrics on the full dataset
accuracy = accuracy_score(y, y_pred)
precision = precision_score(y, y_pred, average='macro', zero_division=1)
recall = recall_score(y, y_pred, average='macro')
f1 = f1_score(y, y_pred, average='macro')

# Print final metrics
print("\nFinal metrics on the full dataset:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-score: {f1:.4f}")

# Plot cumulative confusion matrix with requested adjustments
plt.figure(figsize=(8, 6))

# Custom colormap where lower values are darker
sns.heatmap(cumulative_conf_matrix, annot=True, fmt='d', cmap='Greens_r', 
            cbar=True, xticklabels=np.arange(num_classes), yticklabels=np.arange(num_classes),
            linewidths=0.5, linecolor='black')

plt.title('Cumulative Confusion Matrix Across All Folds')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```
