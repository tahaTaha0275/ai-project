### cp chest pain type
1. Typical Angina
Definition: This is the most common type of chest pain, often described as a squeezing, pressure, or fullness in the center of the chest.
Cause: It occurs when the heart muscle doesn't receive enough oxygen, usually due to narrowed coronary arteries (atherosclerosis).
Symptoms: In addition to chest pain, typical angina can also cause:
Shortness of breath
Fatigue
Sweating
Nausea or vomiting
2. Atypical Angina
Definition: This type of chest pain is similar to typical angina but doesn't follow a predictable pattern.
Cause: It can also be caused by reduced blood flow to the heart muscle but may not be as severe.
Symptoms: Symptoms can vary, but they often include chest discomfort, fatigue, and shortness of breath.
3. Non-Anginal Chest Pain
Definition: This type of chest pain is not caused by heart problems.
Causes: Non-anginal chest pain can be caused by various conditions, including:
Acid reflux (heartburn)
Muscle strain
Costochondritis (inflammation of the cartilage connecting the ribs to the sternum)
Anxiety or panic attacks
Symptoms: Symptoms can vary depending on the underlying cause.
4. Asymptomatic Angina
Definition: This occurs when there is a reduction in blood flow to the heart muscle, but the person experiences no symptoms.
Cause: It can be caused by the same factors as typical angina but without causing noticeable discomfort.

Typical angina, often caused by narrowed coronary arteries, is associated with a higher risk than non-anginal chest pain.

### restecg (resting electrocardiographic results):

This feature represents the results of a resting electrocardiogram (ECG or EKG), a non-invasive test that records the electrical activity of the heart. The results can be categorized into three main categories:

Normal: The ECG shows no abnormalities in the heart's electrical activity.

stt abnormality: There are abnormalities in the ST-segment or T-wave of the ECG, which can be indicative of heart problems such as ischemia (reduced blood flow to the heart) or myocardial infarction (heart attack).

lv hypertrophy: The left ventricle of the heart is enlarged, which can be a sign of high blood pressure or other heart conditions.

### thalach (maximum heart rate achieved):

This feature represents the highest heart rate a person was able to reach during a stress test.
A lower maximum heart rate can be a sign of heart disease, as the heart may not be able to pump blood as efficiently as it should.

### exang (exercise-induced angina):

This feature is a binary variable (True or False) indicating whether the person experienced chest pain or discomfort during exercise.
A value of "True" indicates that the person had exercise-induced angina, while a value of "False" indicates that they did not.
Exercise-induced angina is a common symptom of coronary artery disease.


Experiencing chest pain during exercise is a common symptom of coronary artery disease.

### oldpeak (ST depression induced by exercise relative to rest):

This feature measures the amount of ST-segment depression observed during exercise compared to the resting state.
The ST segment is a portion of the ECG that represents the period between the end of the QRS complex (when the heart muscle contracts) and the beginning of the T wave (when the heart muscle relaxes).
If the ST segment is depressed during exercise, it can be a sign of ischemia (reduced blood flow to the heart muscle). A higher oldpeak value generally indicates a greater degree of ST-segment depression and a higher risk of heart disease.

The amount of ST-segment depression during exercise can indicate the severity of ischemia (reduced blood flow to the heart muscle).

### slope (the slope of the peak exercise ST segment):

This feature describes the shape of the ST segment during exercise.
A downward-sloping ST segment can be a sign of ischemia, while a flat or upsloping ST segment may be less indicative of heart problems.
A steeper downward slope of the ST segment may suggest a more severe degree of ischemia.

The shape of the ST segment during exercise can provide additional clues about the severity of ischemia.

The ST segment is a portion of the electrocardiogram (ECG or EKG) that represents the period between the end of the QRS complex and the beginning of the T wave.

QRS complex: This represents the electrical activity of the heart muscle as it contracts.
T wave: This represents the electrical activity of the heart muscle as it relaxes.
The ST segment is important because it can provide clues about the health of the heart muscle. If the ST segment is elevated or depressed, it can be a sign of heart problems, such as:

Ischemia: Reduced blood flow to the heart muscle
Myocardial infarction: Heart attack
Pericarditis: Inflammation of the pericardium, the sac that surrounds the heart

An elevated ST segment on an electrocardiogram (ECG) is a sign of acute myocardial infarction (heart attack).
A depressed ST segment on an electrocardiogram (ECG) is a sign of ischemia, which means reduced blood flow to the heart muscle.

### ca (number of major vessels colored by fluoroscopy):

This refers to the number of major coronary arteries that are narrowed or blocked, as identified through a medical imaging technique called fluoroscopy.
The number ranges from 0 to 3, indicating the extent of coronary artery disease.

0: No major vessels are narrowed or blocked.

1: One major vessel is narrowed or blocked.

2: Two major vessels are narrowed or blocked.

3: Three major vessels are narrowed or blocked.

### thal (thalassemia test result):

This refers to the results of a thallium stress test, which is used to evaluate the heart's function under stress.
The results can be categorized into three categories:

Normal: The heart muscle receives adequate blood flow even under stress.

Fixed defect: There is a persistent reduction in blood flow to the heart muscle, even at rest.

Reversible defect: The heart muscle receives inadequate blood flow during stress but returns to normal when the stress is relieved.

### Data imputation is a method for retaining the majority of the dataset's data and information by substituting missing data with a different value. These methods are employed because it would be impractical to remove data from a dataset each time. Additionally, doing so would substantially reduce the dataset's size, raising questions about bias and impairing analysis.

### Multivariate imputer that estimates each feature from all the others.

### A strategy for imputing missing values by modeling each feature with missing values as a function of other features in a round-robin fashion.

#### The term "round-robin fashion" in the context of imputing missing values in a dataset refers to a specific iterative approach. Here's a breakdown of how it works:

#### Steps:

##### Identify Features with Missing Values:

###### Determine which features in your dataset have missing values.
##### Select a Feature:

###### Choose one of the features with missing values to start with.

##### Model the Feature:

###### Build a predictive model using the other features in the dataset as predictors to predict the missing values in the selected feature.

##### Impute Missing Values:

###### Use the trained model to predict the missing values in the selected feature.
###### Replace the missing values with the predicted values.
###### Move to the Next Feature:

###### Repeat steps 2-4 for the next feature with missing values, using the updated dataset with the imputed values from the previous steps.

##### Iterate until Complete:

Continue this process, going through each feature with missing values in a circular or "round-robin" manner until all missing values have been imputed.

## Parameters used here
### max_iter int, default=10#### 
Maximum number of imputation rounds to perform before returning the imputations computed during the final round. A round is a single imputation of each feature with missing values. The stopping criterion is met once max(abs(X_t - X_{t-1}))/max(abs(X[known_vals])) < tol, where X_t is X at iteration t. Note that early stopping is only applied if sample_posterior=False.

```python
from sklearn.experimental import enable_iterative_imputer

from sklearn.impute import IterativeImputer

numerical_columns = ['trestbps', 'chol', 'thalch', 'oldpeak', 'age'] 


imputer = IterativeImputer(random_state=42, max_iter=10)

# fitting the iterative imputer on the numeric columns

df[numerical_columns] = imputer.fit_transform(df[numerical_columns])

# Rounding the imputed values to 0 decimal places
df[['trestbps', 'chol', 'thalch', 'age'] ] = df[['trestbps', 'chol', 'thalch', 'age']].round(0)
df['oldpeak'] = df['oldpeak'].round(1)
old_peak = df['oldpeak']
```

```python
from sklearn.impute import KNNImputer
from sklearn.preprocessing import OrdinalEncoder

categorical_cols = ['thal','slope', 'exang', 'restecg', 'fbs', 'cp' , 'sex', 'num'] 
ordinal_enc = OrdinalEncoder()
df[categorical_cols] = ordinal_enc.fit_transform(df[categorical_cols])

imputer = KNNImputer(n_neighbors=10)
df_imputed = imputer.fit_transform(df)

# Convert the imputed data back to DataFrame
df_imputed = pd.DataFrame(df_imputed, columns=df.columns)

# Convert the imputed numerical data back to categorical
df_imputed[categorical_cols] = ordinal_enc.inverse_transform(df_imputed[categorical_cols])
df = df_imputed

# Display the final imputed DataFrame
display(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define a function to plot two features and label the target
def plot_features(df, feature1, feature2, target='num', ax=None):
    sns.scatterplot(data=df, x=feature1, y=feature2, hue=target, ax=ax, palette='deep')
    ax.set_title(f'{feature1} vs {feature2}')
    ax.set_xlabel(feature1)
    ax.set_ylabel(feature2)

# Create 10 subplots
fig, axes = plt.subplots(5, 2, figsize=(15, 25))

# List of feature pairs to plot
feature_pairs = [
    ('age', 'trestbps'),       # Age vs Resting BP TMAM
    ('age', 'chol'),           # Age vs Cholesterol TMAM 
    ('trestbps', 'chol'),      # Resting BP vs Cholesterol TMAM 
    ('thalch', 'oldpeak'),     # Max HR vs ST Depression TMAM
    ('thalch', 'trestbps'),    # Max HR vs Resting BP TMAM
    ('thalch', 'chol'),               # Major Vessels vs Cholesterol NOT TAMAM
    ('age', 'thalch'),         # Age vs Max HR TAMAM
    ('oldpeak', 'trestbps'),        # Major Vessels vs Resting BP NOT TAMAMA
    ('chol', 'oldpeak'),       # Cholesterol vs ST Depression TAMAM
    ('age', 'oldpeak')         # Age vs ST Depression TAMAM
]

# Plot each pair
for ax, (feature1, feature2) in zip(axes.flatten(), feature_pairs):
    plot_features(df, feature1, feature2, ax=ax)

plt.tight_layout()
plt.show()

df.to_csv('heart_disease_uci_formatted_and_cleaned.csv', index=False)
```
